Importing important libraries.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collction and preprocessing.


```python
sonar_data = pd.read_csv("Copy of sonar data.csv" , header = None)
sonar_data.head()
# sonar_data.size
# sonar_data.shape
sonar_data.describe()  # ||Describe -->> show statistical measures of data.
```

```python
sonar_data[60].value_counts()
```

M = Mine

R = Rock


```python
sonar_data.groupby(60).mean()
```

```python
x = sonar_data.drop(columns = 60 , axis = 1 ) # dropping target value from data.
y = sonar_data[60]    # target values storing in var. y.
```

Train and Test Data

```python
x_train, x_test, y_train, y_test  = train_test_split(x, y , test_size = 0.1 , stratify = y , random_state = 1)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

Model Training --->> Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

Model Evaluation


```python
x_train_predictions = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predictions , y_train)
training_data_accuracy       #accuracy on training data
```

```python
x_test_predictions = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predictions , y_test)
test_data_accuracy        # accuracy on test data
```
